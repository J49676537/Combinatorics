```python
import math
import itertools
import random

##################################################
number_of_teams = 5
max_matchups = 999
max_games = 999
##################################################

def nCr(n,r): return round(math.factorial(n) / (math.factorial(r) * math.factorial(n-r)))
def CRnr(n,r): return round(math.factorial(n+r-1) / (math.factorial(r) * math.factorial(n-1)))

teams = {}
for i in range(1, number_of_teams+1):
    if i < 10: teams["0"+str(i)] = 0
    else: teams[str(i)] = 0
print("Teams:", teams)

tables = list(itertools.combinations(teams, 4))
matchups = list(itertools.combinations_with_replacement(teams, 2))
print("Table Combinations:", nCr(number_of_teams,4), "|", len(tables))
print("Matchup Combinations:", CRnr(number_of_teams,2), "|", len(matchups), "| ({})\n".format(len(matchups)-number_of_teams))

#random.shuffle(tables)

team_matchups = {}
for i in range(len(matchups)): team_matchups[str(matchups[i][0])+"&"+str(matchups[i][1])] = 0
print("Team Matchups:", team_matchups, "\n")

schedule = []
for i in range(len(tables)):
    skip = False

    ######################################################################
    pairs = list(itertools.combinations(tables[i], 2))
    for pair in pairs:
        if team_matchups[pair[0]+"&"+pair[1]] == max_matchups:
            skip = True
    
    for j in range(len(tables[i])):
        if teams[tables[i][j]] == max_games:
            skip = True
    ######################################################################

    if skip == False:
        print(tables[i])
        schedule.append(tables[i])
        for j in range(len(tables[i])):
            teams[tables[i][j]] += 1
        for pair in pairs:
            team_matchups[pair[0]+"&"+pair[1]] += 1
    else:
        print(tables[i], "skip")
        continue

print("Teams:", teams)
total_games = 0
for x in teams: total_games += teams[x]
print("Total Games:", round(total_games/4), "\n")

#print("Team Matchups", team_matchups, "\n")

grid = []
for x in team_matchups: grid.append(team_matchups[x])

print("     ", end="")
for i in range(1, number_of_teams+1):
    if i > 9: print(i, sep="", end=" ")
    else: print("0", i, sep="", end=" ")
print("\n     ", "="*number_of_teams*3, sep="")

for i in range(number_of_teams, 0, -1):
    if i < (number_of_teams-8): print("T{}:".format(number_of_teams-i+1), sep="", end="")
    else: print("T0{}:".format(number_of_teams-i+1), sep="", end="")
    indent = "   " * (number_of_teams-i)
    print(indent, end=" ")
    for j in range(i):
        if grid[j] == 0:
            print("--", end=" ")
        elif grid[j] > 9:
            print(grid[j], end=" ")
        else:
            print("0", grid[j], sep="", end=" ")
    print()
```
